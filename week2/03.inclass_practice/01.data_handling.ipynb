##### 데이터 선택

- df['A'] -> 데이터프레임의 A 열 가져오기
- df[['A','B']] -> 여러 열 가져올때는 괄호안에 넣어주면 됨.
- df[0:5] -> 행 기준 슬라이싱

- loc(label of column) -> 열과 행의 조건을 모두 제공함
  * df.loc[:, ['A', 'B']]
- iloc(index/integer location) -> 인덱스의 번호를 사용하여 열람
  * df.iloc[0:10, 1] -> 행, 열 순서 0부터 9까지의 행의 2열을 가지고와라.
- 조건을 이용한 선택
  * df[df['A'] > 0] = df[df.A > 0]
    df.query('A > 0 ') # 잘 안씀

- filter, mask
  * filter : 데이터의 특정 컬럼을 선택하거나, 특정 문자를 포함하는 컬럼 및 데이터 추출
  * mask : 조건에 해당하는 데이터 행을 마스크 처리(조건을 주고, 조건에 해당하는 값을 변경)

merge, join

```python
# 53 Page (데이터결합 연습)
# left, right 데이터 생성
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
result
```

```python
# 57 Page (데이터결합 연습)
# left, right 데이터 생성
left = pd.DataFrame({'key1': [1, 2, 3, 4],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'C': ['C1', 'C2', 'C5'],
                      'D': ['D1', 'D2', 'D5']},
                     index=[1, 2, 5])
```

```python
result = left.join(right) # join의 기본 옵션은 left
result
```

```python
result = left.join(right, how = 'right')
result
```

```python
result = left.join(right, how = 'inner')
result
```

```python
result = left.join(right, how = 'outer')
result
```

```python
result = left.join(right, on = 'key1')
result
```

```python
result = pd.merge(left,right, left_on = 'key1', right_index=True, how = 'left')
result 
```
